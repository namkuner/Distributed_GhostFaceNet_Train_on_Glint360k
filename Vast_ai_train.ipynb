```python
import mxnet as mx
import os
import re
def merge_rec_files(input_dir, output_rec_file):
    part_files = sorted([f for f in os.listdir(input_dir) if f.endswith('.rec')])
    def natural_key(file_name):
      #  Sử dụng biểu thức chính quy để trích xuất số
      return [int(text) if text.isdigit() else text for text in re.split(r'(\d+)', file_name)]

  # Sắp xếp danh sách với khóa tùy chỉnh
    part_files = sorted(part_files, key=natural_key)
    print("part_files",part_files)
    writer = mx.recordio.MXIndexedRecordIO(output_rec_file + '.idx', output_rec_file + '.rec', 'w')
    global_idx = 0

    for part_file in part_files:
        record = mx.recordio.MXIndexedRecordIO(os.path.join(input_dir, part_file.replace('.rec', '.idx')),
                                               os.path.join(input_dir, part_file), 'r')
        for idx in range(len(record.keys)):
            record.seek(idx)
            s = record.read()
            writer.write_idx(global_idx, s)
            global_idx += 1

        record.close()

    writer.close()

# Sử dụng hàm
merge_rec_files('glint360k', 'glint360k/train')
```
